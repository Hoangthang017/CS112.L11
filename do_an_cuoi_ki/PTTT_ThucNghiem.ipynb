##**import thư viện**

```
import pandas as pd
import numpy as np
import math
from sklearn.linear_model import LinearRegression
```

##**Đọc data**

```
data = pd.read_csv('/content/data.csv')
# del data['Unnamed: 2']
# del data['Unnamed: 3']
print(data)
data.columns = ['size', 'operations']

import matplotlib.pyplot as plt
plt.scatter(data['size'], data['operations'])
```

##**Chuẩn hóa dữ liệu**

```
# from google.colab import drive
# drive.mount('/content/drive')
```

```
X = data['size'].values.reshape(-1,1)
```

##**Hàm tìm hệ số x**

```
def Find_X(_X):
  regr = LinearRegression()
  regr.fit(_X, data['operations'])
  return regr.coef_
```

#**Các hàm tìm hệ số x**

```
result = []
```

##**Hàm lg(n)**

```
X_ = np.log2(X)
```

```
re = Find_X(X_)
print(re)
result.append(["lg",re])
```

##**Hàm sqrt(n)**

```
X_ = np.sqrt(X)
```

```
re = Find_X(X_)
print(re)
result.append(["sqrt",re])
```

##**Hàm N**

```
X_ = X
```

```
re = Find_X(X_)
print(re)
result.append(["N",re])
```

##**Hàm n*log(n)**

```
X_ = X*np.log2(X)
```

```
re = Find_X(X_)
print(re)
result.append(["nlgn",re])
```

##**Hàm pow(n,2)**

```
X_ = X**2
```

```
re = Find_X(X_)
print(re)
result.append(["n**2",re])
```

##**Hàm pow(n,3)**

```
X_ = X**3
```

```
re = Find_X(X_)
print(re)
result.append(["n**3",re])
```

##**Hàm pow(2,n)**

```
X_ = 2**X
```

```
re = Find_X(X_)
print(re)
result.append(re)
```
